```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url ='https://ekantipur.com'
response = requests.get(url)

soup = BeautifulSoup(response.content, 'html')
```

```python
print(soup)
```

```python
article= soup.find_all('article')
```

```python
print(len(article))
```

```python
article[10]
```

```python
data = {
    'headlines': [],
    'body': [],
}
df = pd.DataFrame(data)
df
```

```python
for i in article:
    headlines= i.find('h2')
    body = i.find('p').text
    cleanedBody = body.replace('\xa0', ' ')
    df.loc[len(df)] = [headlines, cleanedBody]


df

    
```
